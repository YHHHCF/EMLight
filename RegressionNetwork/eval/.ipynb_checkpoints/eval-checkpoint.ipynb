```python
import sys
sys.path.insert(1, '../')
import util
import matplotlib.pyplot as plt
```

```python
handle = util.PanoramaHandler()
tone = util.TonemapHDR()
```

```python
def ldr_from_hdr(path):
    hdr = handle.read_hdr(path)
    ldr = tone(hdr, True)[0]
    return ldr
```

```python
original_map = ldr_from_hdr('AG8A8645-e4908f5e6f.exr')
warped_map = ldr_from_hdr('AG8A8645-others-40-1.79482-0.98343.exr')
input = ldr_from_hdr('AG8A8645-others-40-1.79482-0.98343_input.exr')
gt_light = handle.read_hdr('AG8A8645-others-40-1.79482-0.98343_gt.exr')
pred_light = handle.read_hdr('AG8A8645-others-40-1.79482-0.98343_pred.exr')
```

```python
plt.imshow(original_map)
plt.title(("original_map", original_map.shape))
plt.show()

plt.imshow(warped_map)
plt.title(("warped_map", warped_map.shape))
plt.show()

plt.imshow(input)
plt.title(("input", input.shape))
plt.show()

plt.imshow(gt_light)
plt.title(("gt_light", gt_light.shape))
plt.show()

plt.imshow(pred_light)
plt.title(("pred_light", pred_light.shape))
plt.show()
```

```python
import cv2
pred_diffuse = cv2.cvtColor(cv2.imread('pred_diffuse.png'), cv2.COLOR_BGR2RGB)
pred_mirror = cv2.cvtColor(cv2.imread('pred_mirror.png'), cv2.COLOR_BGR2RGB)
gt_diffuse = cv2.cvtColor(cv2.imread('gt_diffuse.png'), cv2.COLOR_BGR2RGB)
gt_mirror = cv2.cvtColor(cv2.imread('gt_mirror.png'), cv2.COLOR_BGR2RGB)
gt_env_diffuse = cv2.cvtColor(cv2.imread('gt_env_diffuse.png'), cv2.COLOR_BGR2RGB)
gt_env_mirror = cv2.cvtColor(cv2.imread('gt_env_mirror.png'), cv2.COLOR_BGR2RGB)

f, axs = plt.subplots(2, 3, figsize=(15, 8))
axs[0, 0].set_title(("pred_diffuse", pred_diffuse.shape))
axs[0, 0].imshow(pred_diffuse)
axs[0, 1].set_title(("gt_diffuse", gt_diffuse.shape))
axs[0, 1].imshow(gt_diffuse)
axs[0, 2].set_title(("gt_env_diffuse", gt_env_diffuse.shape))
axs[0, 2].imshow(gt_env_diffuse)
axs[1, 0].set_title(("pred_mirror", pred_mirror.shape))
axs[1, 0].imshow(pred_mirror)
axs[1, 1].set_title(("gt_mirror", gt_mirror.shape))
axs[1, 1].imshow(gt_mirror)
axs[1, 2].set_title(("gt_env_mirror", gt_env_mirror.shape))
axs[1, 2].imshow(gt_env_mirror)
plt.show()
```

```python
import numpy as np
mask_num = np.sum(pred_diffuse > 0)
RMSE_diffuse = np.sqrt(np.sum((pred_diffuse - gt_diffuse) ** 2) / mask_num)
RMSE_mirror = np.sqrt(np.sum((pred_mirror - gt_mirror) ** 2) / mask_num)
print(RMSE_diffuse, RMSE_mirror)
```
